- optimization problem <br/>
$$
\begin{align}
& minimize \qquad f_0(x)\\
& subject\enspace to \qquad f_i(x) \leq b_i,\enspace i = 1,...,m\\
& x = (x_1, ..., x_n)\\
& f_0: R^n \rightarrow R\\
& f_i: R^n \rightarrow R, \enspace i = 1,...,m\\
&arg\,min\;f_0(x) := x^*
\end{align}
$$
- convex optimization problem
the constraints satisfy

$$
\begin{align}
&f_i(\alpha x + \beta y ) \leq \alpha f_i(x) + \beta f_i(y), \\
&\alpha + \beta = 1, \alpha \geq 0, \beta \geq 0
\end{align}
$$
